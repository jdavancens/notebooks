<h2>(1) linear interpolation</h2>

linear interpolation between two values $v_1$ and $v_2$:<br><br>
<h2>&nbsp;&nbsp;&nbsp;&nbsp; $v_t = v_1 + (v_2 - v_1) t$<br><br></h2>
where $t \in [0,1]$ and $v_t = v_1$ for $t = 0$ and $v_t = v_2$ for $t = 1$

```python
# linear intepolation between two values
def lin_interp(t, v1, v2): 
    return (float(v2-v1)*t)+v1



# some examples
print("some examples of finding an interpolated value between two points")
print lin_interp(0.0, 3, 10)
print lin_interp(1.0, 0, 1)
print lin_interp(0.5, 3.7, 4.2)
print lin_interp(0.2, 11, 11.002)
print lin_interp(0.7, 3, 4)

# plot

v1 = 0
v2 = 1
num_steps = 40
ts = [i/float(num_steps) for i in range(num_steps+1)]
vs = [lin_interp(t, v1, v2) for t in ts]
print("\n\n\nsimple linear interpolation between two points")
plt.scatter([v1,v2], [0,1])
plt.plot(ts, vs, c = "R", linestyle = "--");

```

```python
# break-point linear interpolation on a list of values (given above)
def lin_interp_breakpoint(t, values):
    v1_index = int(math.floor(t))
    v2_index = int(math.ceil(t))
    return lin_interp(t-math.floor(t), values[v1_index], values[v2_index])

# plot original
plt.figure(figsize=(12,6))


# plot interpolation
# note, the plot program could have done this
num_steps = 40
ts = [i/float(num_steps)*(len(values)-1) for i in range(num_steps+1)]
vs = [lin_interp_breakpoint(t, values) for t in ts]
plt.scatter(ts,vs, c = "cyan")
plt.scatter(range(len(values)), values, c = "magenta");
# plt.plot(ts,vs);
```

<h2>(2) half cosine interpolation</h2>

half cosine interpolation between two values $v_1$ and $v_2$:<br><br>
<h2>&nbsp;&nbsp;&nbsp;&nbsp; $v_t = \frac{v_1+v_2}{2} + \frac{v_1-v_2}{2} \cos(\pi t)$<br><br></h2>
<h3>where $t \in [0,1]$ and $v_t = v_1$ for $t = 0$ and $v_t = v_2$ for $t = 1$</h3>

```python
# half cosine intepolation between two points
def hcos_interp(t, v1, v2): 
    return (v1+v2)/2.0 + ((v1-v2)/2.0)*math.cos(math.pi*t)
    
# some examples
print hcos_interp(0.0, 3, 10)
print hcos_interp(1.0, 0, 1)
print hcos_interp(0.5, 3.7, 4.2)
print hcos_interp(0.2, 11, 11.002)
print hcos_interp(0.7, 3, 4)

# plot
v1 = 0
v2 = 1
num_steps = 40
ts = [i/float(num_steps) for i in range(num_steps+1)]
vs = [hcos_interp(t, v1, v2) for t in ts]
plt.scatter([v1,v2], [0,1])
plt.plot(ts, vs);
```

```python
# break-point half cosine interpolation on a list of values
def hcos_interp_breakpoint(t, values):
    v1_index = int(math.floor(t))
    v2_index = int(math.ceil(t))
    return hcos_interp(t-math.floor(t), values[v1_index], values[v2_index])

# plot original
plt.figure(figsize=(12,6))
plt.scatter(range(len(values)), values);

# plot interpolation
num_steps = 80
ts = [i/float(num_steps)*(len(values)-1) for i in range(num_steps+1)]
vs = [hcos_interp_breakpoint(t, values) for t in ts]
plt.plot(ts, vs);
```

<h2>(3) exponential interpolation</h2>

exponential interpolation between two values $v_1$ and $v_2$:<br><br>
<h1>&nbsp;&nbsp;&nbsp;&nbsp; $v_t = v_1 + \frac{e^{kt}-1}{e^k-1} (v_2-v_1)$<br><br></h1>
<h2>where $t \in [0,1]$ and $v_t = v_1$ for $t = 0$ and $v_t = v_2$ for $t = 1$</h2>

```python
# exponential intepolation between two points
# implements exponential growth when v1 < v2
# implements exponential decay (negative k) when v1 > v2
def exp_interp(t, v1, v2, k=1):
    k = k if v2>v1 else -k
    return v1 + (math.e**(float(k)*t)-1)/(math.e**float(k)-1) * (v2-v1)

# some examples
print exp_interp(0.0, 3, 10)
print exp_interp(1.0, 0, 1)
print exp_interp(0.5, 3.7, 4.2)
print exp_interp(0.2, 11, 11.002)
print exp_interp(0.7, 3, 4)

# plot
v1 = 0
v2 = 1
num_steps = 100
ts = [i/float(num_steps) for i in range(num_steps+1)]
vs = [exp_interp(t, v1, v2, k=1) for t in ts]
plt.scatter([v1,v2], [0,1])
plt.plot(ts, vs);
```

```python
# the exponential functions has an additional parameter, k

# params
k_min = -20
k_max = 20
k_num = 10

# plots a curve for a given k
def plot(k):
    v1 = 0
    v2 = 1
    num_steps = 40
    ts = [i/float(num_steps) for i in range(num_steps+1)]
    vs = [exp_interp(t, v1, v2, k=k) for t in ts]
    plt.plot(ts, vs);
    
# plot them all
plt.figure(figsize = (9,5))
for i in range(k_num+1):
    k = i*(k_max-k_min)/float(k_num) + k_min
    if k != 0.0: plot(k)  # k=0 is constant (does not interpolate)
```

```python
# break-point exponential interpolation on a list of values, k=4
def exp_interp_breakpoint(t, values, k=1):
    v1_index = int(math.floor(t))
    v2_index = int(math.ceil(t))
    return exp_interp(t-math.floor(t), values[v1_index], values[v2_index], k=k)

# plot original
plt.figure(figsize=(12,6))
plt.scatter(range(len(values)), values);

# plot interpolation
num_steps = 80
ts = [i/float(num_steps)*(len(values)-1) for i in range(num_steps+1)]
vs = [exp_interp_breakpoint(t, values, k=4) for t in ts]
plt.plot(ts, vs);
```

<h2>example: interpolation between morphologies</h2>

```python
# example: interpolation between morphologies

# generate melodies
random.seed(6)
num_points = 6
source = zip(sorted([0] + [random.random() for i in range(num_points-2)] + [1]), [random.random() for i in range(num_points)])
target = zip(sorted([0] + [random.random() for i in range(num_points-2)] + [1]), [random.random() for i in range(num_points)])

# plot source and target
plt.figure()
plt.title('source (green) and target (blue)')
plt.plot([p[0] for p in source], [p[1] for p in source], c='g');
plt.plot([p[0] for p in target], [p[1] for p in target], c='b');

# general contour interpolate
def interp_contour(source, target, func, i):
    return [(func(i, node[0][0], node[1][0]), func(i, node[0][1], node[1][1])) for node in zip(source, target)]

# linear interpolate!
lin_interpolated = [interp_contour(source, target, lin_interp, step/10.0) for step in range(1,10)]
plt.figure()
plt.title('linear')
plt.plot([p[0] for p in source], [p[1] for p in source], c='g');
plt.plot([p[0] for p in target], [p[1] for p in target], c='b');
for inter in lin_interpolated:
    plt.plot([p[0] for p in inter], [p[1] for p in inter], c='r');

# hcos interpolate!
hcos_interpolated = [interp_contour(source, target, hcos_interp, step/10.0) for step in range(1,10)]
plt.figure()
plt.title('half cosine')
plt.plot([p[0] for p in source], [p[1] for p in source], c='g');
plt.plot([p[0] for p in target], [p[1] for p in target], c='b');
for inter in hcos_interpolated:
    plt.plot([p[0] for p in inter], [p[1] for p in inter], c='r');
    
# exp interpolate!
exp_interpolated = [interp_contour(source, target, exp_interp, step/10.0) for step in range(1,10)]
plt.figure()
plt.title('exponential')
plt.plot([p[0] for p in source], [p[1] for p in source], c='g');
plt.plot([p[0] for p in target], [p[1] for p in target], c='b');
for inter in exp_interpolated:
    plt.plot([p[0] for p in inter], [p[1] for p in inter], c='r');

# # compare all three
for i in range(9):
    step = (i+1)/10.0
    plt.figure()
    plt.title('i = ' + str(step))
    plt.plot([p[0] for p in lin_interpolated[i]], [p[1] for p in lin_interpolated[i]], c='g');
    plt.plot([p[0] for p in hcos_interpolated[i]], [p[1] for p in hcos_interpolated[i]], c='b');
    plt.plot([p[0] for p in exp_interpolated[i]], [p[1] for p in exp_interpolated[i]], c='r');
```

```python
# now let's hear it

def melody_2_midi(interpolated, filename):
    
    # create a pm object
    pm = pretty_midi.PrettyMIDI()
    pm.instruments.append(pretty_midi.Instrument(0))

    # midi parameteres
    pitch_low = 48
    pitch_hi = 72
    time_scale = 1  # melody length
    dur = 0.3
    velocity = 100

    # stuff the pm object
    for i,inter in enumerate(interpolated):
        for note in inter:
            raw_time, raw_pitch = note
            pitch = int(round((pitch_hi-pitch_low)*raw_pitch+pitch_low))
            note_on = i*(time_scale+.2) + raw_time*time_scale
            note_off = note_on+dur
            pm.instruments[0].events.append(pretty_midi.Note(velocity, pitch, note_on, note_off))

    # and write it
    pm.write(filename)
    
melody_2_midi(lin_interpolated, 'fomulas_interpolation_linear_melodies.mid')
melody_2_midi(hcos_interpolated, 'fomulas_interpolation_hcos_melodies.mid')
melody_2_midi(exp_interpolated, 'fomulas_interpolation_exp_melodies.mid')
```

<h2>example: 2D interpolation between points</h2>

```python
# example: 2D interpolation between points

# first some random points
random.seed(3)
num_points = 6
xs = sorted([0] + [random.random() for i in range(num_points-2)] + [1])
ys = [random.random() for i in range(num_points)]
points2D = zip(xs, ys)

# interpolation parameters
num_steps = 80
xfunc = exp_interp  # we can have different interp funcs for x and y dims
yfunc = hcos_interp  # we can have different interp funcs for x and y dims

# function def
def interp_breakpoint_2D(t, points, xfunc, yfunc):

    p1_index = int(math.floor(t))
    p2_index = int(math.ceil(t))

    def interp_2D(t, p1, p2, xfun, yfun):
        xt = xfun(t, 0.0, 1.0)
        x = xfun(t, p1[0], p2[0])
        y = yfun(xt, p1[1], p2[1])
        return (x, y)

    return points[p1_index] if p1_index is p2_index else interp_2D(t-math.floor(t), points[p1_index], points[p2_index], xfunc, yfunc)

# plot original
plt.scatter([p[0] for p in points2D], [p[1] for p in points2D]);    

# plot interpolation
ilist = [i/float(num_steps)*(len(points2D)-1) for i in range(num_steps+1)]
interped = [interp_breakpoint_2D(i, points2D, xfunc, yfunc) for i in ilist]
plt.figure(figsize=(12,6))
plt.scatter([p[0] for p in interped], [p[1] for p in interped], s=5);
```

```python
# now let's hear it

# create a pm object
pm = pretty_midi.PrettyMIDI()
pm.instruments.append(pretty_midi.Instrument(0))

# midi parameteres
pitch_low = 36
pitch_hi = 104
time_scale = 30
dur = 0.4
velocity = 100

# stuff the pm object
for note in interped:
    
    raw_time, raw_pitch = note
    pitch = int(round((pitch_hi-pitch_low)*raw_pitch+pitch_low))
    note_on = raw_time*time_scale
    note_off = note_on+dur
    pm.instruments[0].events.append(pretty_midi.Note(velocity, pitch, note_on, note_off))
    
    # print pitch, note_on, note_off
    
# and write it
pm.write("fomulas_interpolation_ex2.mid")
```
